```python
import pandas as pd
```

```python
crop_df = pd.read_csv("crop_production_data.csv")
crop_df.head()
```

```python
crop_df.columns = crop_df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
crop_df.head()
```

```python
crop_df.isnull().sum()
```

```python
crop_df = crop_df.dropna(subset=['production_'])
crop_df = crop_df[crop_df['production_'] > 0]
crop_df = crop_df[crop_df['area_'] > 0]
```

```python
crop_df.isnull().sum()
```

```python
crop_df = (
    crop_df.groupby(['state_name', 'crop_year', 'season', 'crop'], as_index=False)
    .agg({'area_': 'sum', 'production_': 'sum'})
)

crop_df['yield_'] = crop_df['production_'] / crop_df['area_']

crop_df = crop_df.sort_values(by=['state_name', 'crop_year', 'season', 'crop']).reset_index(drop=True)

crop_df.head()
```

```python
print("Unique States:", crop_df['state_name'].nunique())
print(crop_df['state_name'].unique())

print("\nUnique Seasons:", crop_df['season'].nunique())
print(crop_df['season'].unique())

print("\nUnique Crops:", crop_df['crop'].nunique())
print(crop_df['crop'].unique())
```

```python
print("Year range:", crop_df['crop_year'].min(), "to", crop_df['crop_year'].max())
```

```python
rain_df = pd.read_csv("rainfall_data.csv")
rain_df.head()
```

```python
rain_df = rain_df[(rain_df['YEAR'] >= 1997) & (rain_df['YEAR'] <= 2015)]
```

```python
rain_df = rain_df.drop(columns=['JF', 'MAM', 'JJAS', 'OND'], errors='ignore')
```

```python
rain_df.isnull().sum()
```

```python
rain_df.head()
```

```python
rain_df['ANNUAL'] = rain_df[['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']].sum(axis=1)
print(rain_df['ANNUAL'].isnull().sum())
rain_df.head()
```

```python
rain_df = rain_df.fillna(0)
```

```python
rain_df = rain_df.rename(columns={
    'SUBDIVISION': 'state_name',
    'YEAR': 'crop_year'
})
```

```python
rain_df.head()
```

```python
print(rain_df['state_name'].unique())
```

```python
multi_state_groups = [
    'Assam & Meghalaya',
    'Naga Mani Mizo Tripura',
    'Sub Himalayan West Bengal & Sikkim',
    'Haryana Delhi & Chandigarh',
    'Konkan & Goa',
    'Coastal Andhra Pradesh',
    'Rayalseema',
    'Coastal Karnataka',
    'North Interior Karnataka',
    'South Interior Karnataka'
]

# Drop only those rows
rain_df_cleaned = rain_df[~rain_df['state_name'].isin(multi_state_groups)].copy()
print(rain_df_cleaned['state_name'].unique())
```

```python
rename_to_state = {
    'Andaman & Nicobar Islands': 'Andaman and Nicobar Islands',
    'Orissa': 'Odisha',
    'East Uttar Pradesh': 'Uttar Pradesh',
    'West Uttar Pradesh': 'Uttar Pradesh',
    'West Rajasthan': 'Rajasthan',
    'East Rajasthan': 'Rajasthan',
    'West Madhya Pradesh': 'Madhya Pradesh',
    'East Madhya Pradesh': 'Madhya Pradesh',
    'Gujarat Region': 'Gujarat',
    'Saurashtra & Kutch': 'Gujarat',
    'Madhya Maharashtra': 'Maharashtra',
    'Matathwada': 'Maharashtra',
    'Vidarbha': 'Maharashtra',
    'Gangetic West Bengal': 'West Bengal',
    'Jammu & Kashmir': 'Jammu and Kashmir'
}

rain_df_cleaned['state_name'] = rain_df_cleaned['state_name'].replace(rename_to_state)
```

```python
rain_df_yearly = (
    rain_df_cleaned
    .groupby(['state_name', 'crop_year'], as_index=False)
    .mean()
)
```

```python
print(rain_df_yearly['state_name'].unique())
```

```python
rain_states = set(rain_df_yearly['state_name'].unique())
crop_states = set(crop_df['state_name'].unique())

common_states = rain_states.intersection(crop_states)

rain_df_final = rain_df_yearly[rain_df_yearly['state_name'].isin(common_states)].copy()
crop_df_final = crop_df[crop_df['state_name'].isin(common_states)].copy()

rain_only = sorted(rain_states - crop_states)
crop_only = sorted(crop_states - rain_states)

print("Common states:", sorted(common_states))
print("In rain_df only:", rain_only)
print("In crop_df only:", crop_only)
```

```python
rain_df_final['Kharif'] = rain_df_final[['JUN', 'JUL', 'AUG', 'SEP']].sum(axis=1)
rain_df_final['Rabi'] = rain_df_final[['OCT', 'NOV', 'DEC', 'JAN', 'FEB', 'MAR']].sum(axis=1)
rain_df_final['Summer'] = rain_df_final[['APR', 'MAY']].sum(axis=1)
rain_df_final['Winter'] = rain_df_final[['JAN', 'FEB']].sum(axis=1)
rain_df_final['Autumn'] = rain_df_final[['OCT', 'NOV']].sum(axis=1)
```

```python
month_cols = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']

rain_df_final = rain_df_final.drop(columns=month_cols)
```

```python
rain_df_final.head()
```

```python
rain_df_final = rain_df_final.rename(columns={'ANNUAL': 'Whole Year'})
```

```python
rain_df_final.head()
```

```python
crop_df_final.head()
```

```python
rain_seasonal = rain_df_final.melt(
    id_vars=['state_name', 'crop_year'],
    value_vars=['Whole Year', 'Kharif', 'Rabi', 'Summer', 'Winter', 'Autumn'],
    var_name='season',
    value_name='rainfall_mm'
)
```

```python
merged_df = crop_df_final.merge(
    rain_seasonal,
    on=['state_name', 'crop_year', 'season'],
    how='inner'
)
print(merged_df.head())
print(merged_df.shape)
```

```python
merged_df.to_csv('final_dataset.csv', index=False)
```
